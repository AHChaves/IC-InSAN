## Distribuição das atividades economicas por UF

```python
atividade_por_uf = dados.groupby(['Sigla_UF', 'Atividade com maior valor adicionado bruto']).size().unstack()

# Configurar o estilo do gráfico
plt.figure(figsize=(16, 9))
sns.set_style("whitegrid")

# Número de categorias
num_categorias = len(atividade_por_uf.columns)

# Gerar tons de cinza válidos (de escuro para claro)
cinzas = plt.cm.Greys(np.linspace(0.2, 0.8, num_categorias))  # Evitando preto puro e branco puro

# Padrões de hachura
hatches = ['', '///', '\\\\\\', '|||', '+++', 'xxx', 'ooo', '...', '***'][:num_categorias]

# Criar gráfico de barras empilhadas
ax = atividade_por_uf.plot(kind='bar', 
                         stacked=True, 
                         color=cinzas,
                         edgecolor='black', 
                         linewidth=0.8,
                         width=0.85,figsize=(16,9))

# Adicionar padrões de hachura para diferenciar as categorias
for i, patch in enumerate(ax.patches):
    if i < len(hatches) * len(atividade_por_uf):  # Garantir que não excedemos os padrões
        patch.set_hatch(hatches[i % len(hatches)])

# Adicionar títulos e labels
plt.title('Atividades Econômicas com Maior Valor Adicionado por UF', 
         fontsize=14, fontweight='bold', pad=20, color='black')
plt.xlabel('Unidade Federativa (UF)', fontsize=12, color='black')
plt.ylabel('Número de Municípios', fontsize=12, color='black')
plt.xticks(rotation=45, color='black')
plt.yticks(color='black')

# Configurar legenda
legenda = plt.legend(title='Atividade Principal', 
                #    bbox_to_anchor=(1.05, 1), 
                   loc='upper left',
                   frameon=True,
                   edgecolor='black')
legenda.get_title().set_color('black')
for text in legenda.get_texts():
    text.set_color('black')

# Melhorar a grade
plt.grid(axis='y', color='gray', linestyle=':', alpha=0.7)

# Ajustar layout
plt.subplots_adjust(right=0.8, bottom=0.15, top=0.9)
plt.tight_layout()
plt.savefig("../imagens/AtvEco_UF.png", dpi=600)
plt.show()
```

```python
colunas_producao = [
    "Produção de Soja (em grão)(toneladas)",
    "Produção de Milho (em grão)(toneladas)",
    "Produção de Café (em grão) Total(toneladas)",
    "qtd_familias_total_bolsa_familia",
    "num_beneficiados_bpc",
    "Produção de Arroz (em casca)(toneladas)",
    "Produção de Feijão (em grão)(toneladas)",
    "Produção de Mandioca(toneladas)"
]

for coluna in colunas_producao:
    dados_todos[coluna] = dados_todos[coluna].replace("-", "0")
    dados_todos[coluna] = dados_todos[coluna].replace("...", "0")
    dados_todos[coluna] = dados_todos[coluna].astype(float)
```
