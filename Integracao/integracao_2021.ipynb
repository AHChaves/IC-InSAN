# Integração

```python
bd_2021 = cod_municipios[['Codigo_Municipio', 'Codigo_Municipio_2', 'Município', 'UF', 'Sigla_UF']]
```

```python
bd_2021 = bd_2021.merge(caged, how='left')
```

```python
bd_2021 = bd_2021.merge(pib_Munic[['Codigo_Municipio', 'PIB do Municipio']], how='left')
```

```python
bd_2021 = bd_2021.merge(pib_Munic_BD_2021, how='left')
```

```python
bd_2021 = bd_2021.merge(cempre, how='right')
```

```python
bd_2021 = bd_2021.merge(bolsa_familia, how='left', left_on='Codigo_Municipio_2', right_on='ibge')
```

```python
bd_2021 = bd_2021.drop('ibge', axis=1)
```

```python
bd_2021 = bd_2021.merge(bpc, how='left', left_on='Codigo_Municipio_2', right_on='ibge')
```

```python
bd_2021 = bd_2021.drop('ibge', axis=1)
```

```python
bd_2021 = bd_2021.merge(clin_Amb_BAS, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021 = bd_2021.merge(clin_Amb_ESP, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
```

```python
bd_2021 = bd_2021.merge(leitos_internacao, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
```

```python
bd_2021.drop(columns=['cod_municipio', 'cod_municipio_x', 'cod_municipio_y'], inplace=True)
```

```python
bd_2021 = bd_2021.merge(leitos_repouso_Ambulatorio_FEM, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021 = bd_2021.merge(leitos_repouso_Ambulatorio_MASC, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021 = bd_2021.merge(leitos_repouso_Ambulatorio_INDF, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021.drop(columns=['cod_municipio', 'cod_municipio_x', 'cod_municipio_y'], inplace=True)
bd_2021 = bd_2021.merge(leitos_repouso_Ambulatorio_PED, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
```

```python
bd_2021 = bd_2021.merge(leitos_repouso_Urgencia_FEM, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021 = bd_2021.merge(leitos_repouso_Urgencia_MASC, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021.drop(columns=['cod_municipio', 'cod_municipio_x', 'cod_municipio_y'], inplace=True)
bd_2021 = bd_2021.merge(leitos_repouso_Urgencia_INDF, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021 = bd_2021.merge(leitos_repouso_Urgencia_PED, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2021.drop(columns=['cod_municipio_x', 'cod_municipio_y'], inplace=True)
```

```python
bd_2021 = bd_2021.merge(PAM_2021, how='left')
```
