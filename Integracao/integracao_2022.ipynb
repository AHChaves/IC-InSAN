### Esgoto

```python
esgoto = esgoto[['Codigo_Municipio', 'Total', 'Rede geral, rede pluvial ou fossa ligada à rede']]
```

```python
esgoto.rename(columns={"Total":"Total de domicilios com acesso a rede de esgoto", "Rede geral, rede pluvial ou fossa ligada à rede":"Número de domicilios com esgoto coletado pela rede geral, rede pluvial ou fossa ligada à rede"}, inplace=True)
```

### Distribuição de Agua

```python
d_agua = pd.DataFrame()
d_agua['Codigo_Municipio'] = dist_agua['Codigo_Municipio '].dropna(inplace=False)
i = 0

d_agua.reset_index(inplace=True)
d_agua.drop(columns=['index'], inplace=True)

for index, row in dist_agua.iterrows():
    if ('Canalizada até dentro da casa, apartamento ou habitação' in row['Unnamed: 2']) and (dist_agua.at[index, 'Codigo_Municipio '] == d_agua.at[i, 'Codigo_Municipio']):
        d_agua.at[i, 'Total domicilios com acesso a água, até dentro da casa, apartamento ou habitação'] = dist_agua.at[index, 'Total']
        d_agua.at[i, 'Numero de domicilios com acesso a água, até dentro da casa, apartamento ou habitação, pela rede geral de distribuição'] = dist_agua.at[index,'Rede geral de distribuição']
    elif 'Canalizada, mas apenas no terreno' in row['Unnamed: 2'] and dist_agua.at[index - 1, 'Codigo_Municipio '] == d_agua.at[i, 'Codigo_Municipio']:
        d_agua.at[i, 'Total domicilios com acesso a água, mas apenas no terreno'] = dist_agua.at[index,'Total']
        d_agua.at[i, 'Numero de domicilios com acesso a água, mas apenas no terreno, pela rede geral de distribuição'] = dist_agua.at[index,'Rede geral de distribuição']
    elif 'Sem água canalizada' in row['Unnamed: 2'] and dist_agua.at[index - 2, 'Codigo_Municipio '] == d_agua.at[i, 'Codigo_Municipio']:
        d_agua.at[i, 'Sem água canalizada'] = dist_agua.at[index,'Total']
        i = i+1
```

```python
bd_2022 = cod_municipios[['Codigo_Municipio', 'Codigo_Municipio_2', 'Município', 'UF', 'Sigla_UF']]
```

```python
bd_2022 = bd_2022.merge(tipo_domi, how='left')
```

```python
bd_2022 = bd_2022.merge(d_agua, how='left')
```

```python
bd_2022 = bd_2022.merge(caged, how='left')
```

```python
bd_2022 = bd_2022.merge(cempre, how='left')
```

```python
bd_2022 = bd_2022.merge(bpc, how='left', left_on='Codigo_Municipio_2', right_on='ibge')
```

```python
bd_2022 = bd_2022.drop('ibge', axis=1)
```

```python
bd_2022 = bd_2022.merge(clin_Amb_BAS, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022 = bd_2022.merge(clin_Amb_ESP, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
```

```python
bd_2022 = bd_2022.merge(leitos_internacao, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
```

```python
bd_2022.drop(columns=['cod_municipio', 'cod_municipio_x', 'cod_municipio_y'], inplace=True)
```

```python
bd_2022 = bd_2022.merge(leitos_repouso_Ambulatorio_FEM, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022 = bd_2022.merge(leitos_repouso_Ambulatorio_MASC, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022 = bd_2022.merge(leitos_repouso_Ambulatorio_INDF, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022.drop(columns=['cod_municipio', 'cod_municipio_x', 'cod_municipio_y'], inplace=True)
bd_2022 = bd_2022.merge(leitos_repouso_Ambulatorio_PED, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
```

```python
bd_2022 = bd_2022.merge(leitos_repouso_Urgencia_FEM, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022 = bd_2022.merge(leitos_repouso_Urgencia_MASC, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022.drop(columns=['cod_municipio', 'cod_municipio_x', 'cod_municipio_y'], inplace=True)
bd_2022 = bd_2022.merge(leitos_repouso_Urgencia_INDF, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022 = bd_2022.merge(leitos_repouso_Urgencia_PED, how='left',left_on='Codigo_Municipio_2', right_on='cod_municipio')
bd_2022.drop(columns=['cod_municipio_x', 'cod_municipio_y'], inplace=True)
```
