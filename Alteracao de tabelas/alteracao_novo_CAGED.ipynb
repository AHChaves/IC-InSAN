```python
caged = pd.read_excel("../dataset/Economico/Renda e Desenvolvimento/PIB/novo_CAGED_alterado.xlsx", sheet_name="Tabela 8", header=[0,1])
```

```python
cod_municipios = pd.read_csv('../dataset/Código dos Municípios')
```

```python
caged.columns = ['_'.join(col) for col in caged.columns]
```

```python
caged = caged.rename(columns={'Unnamed: 0_level_0_UF': 'UF', 'Unnamed: 1_level_0_Código do Município': 'Codigo do Municipio', 'Unnamed: 2_level_0_Município': 'Municipio'})
```

```python
nomes= []
ufs = []
i = 0

for index, row in caged.iterrows():
    if '-' in row['Municipio']:
        tmp = row['Municipio'].split("-")
        nomes.append(tmp[1])

for index, row in caged.iterrows():
    if '-' in row['Municipio']:
        caged.iat[i, 2] = nomes[i]
        i = i+1
```

```python
caged = caged.drop(index=5570)
```

```python
tmp = list(caged.columns[3:252])
```

```python
cmp_Cols = []
for x in tmp:
    cmp_Cols.append(x.split("/")[1])
```

```python
cmp_Cols_Idx = pd.Index(cmp_Cols)
```

```python
cmp_Cols = list(cmp_Cols_Idx.value_counts().index)
```

```python
estoque_cols = [col for col in cmp_Cols if "Estoque" in col]
fixed_cols = ['UF', 'Codigo do Municipio', 'Municipio']
estoque_columns = fixed_cols + [col for col in caged.columns if any(estoque in col for estoque in estoque_cols)]
```

```python
caged_estoque = caged[estoque_columns]
```

```python
caged_estoque.replace('---', np.nan, inplace=True)
```

```python
media_estoque_por_ano = pd.DataFrame()
for estoque_col in estoque_cols:
    # Filtrando as colunas que contêm o ano atual
    colunas_ano = [col for col in caged_estoque.columns if estoque_col in col]
    
    # Calculando a média das colunas do ano atual
    media_estoque_por_ano[estoque_col + "_media"] = caged_estoque[colunas_ano].mean(axis=1)
    media_estoque_por_ano[estoque_col + "_MAX"] = caged_estoque[colunas_ano].max(axis=1)
    media_estoque_por_ano[estoque_col + "_MIN"] = caged_estoque[colunas_ano].min(axis=1)
    

# Adicionando as colunas fixas ao novo DataFrame
media_estoque_por_ano = pd.concat([caged_estoque[fixed_cols], media_estoque_por_ano], axis=1)
```

```python
cod_municipios["Município"] = cod_municipios["Município"].apply(lambda x: unidecode(str(x)))
```

```python
cod_municipios["Município"] = cod_municipios["Município"].str.replace("'"," ", regex=False)
```

```python
media_estoque_por_ano = media_estoque_por_ano.merge(cod_municipios[['Codigo_Municipio','Sigla_UF','Município']], how='right', left_on=["UF", "Municipio"], right_on=["Sigla_UF", "Município"])
```

```python
media_estoque_por_ano = media_estoque_por_ano.drop(columns=["Municipio", "UF"])
```

```python
media_estoque_por_ano = media_estoque_por_ano.rename(columns={'Codigo do Municipio': 'Codigo_Municipio_2', 'Município': 'Municipio', 'Sigla_UF': 'UF'})
```

```python
media_estoque_por_ano
```

```python
media_estoque_por_ano["Codigo_Municipio"] = media_estoque_por_ano["Codigo_Municipio"].astype(int)
```

```python
media_estoque_por_ano['Codigo_Municipio_2'] = media_estoque_por_ano['Codigo_Municipio_2'].fillna(0).replace([np.inf, -np.inf], 0).astype(int)
```

```python
media_estoque_por_ano = media_estoque_por_ano[['Codigo_Municipio', 'Codigo_Municipio_2', 'UF', 'Municipio'] + [col for col in media_estoque_por_ano.columns if col != 'Codigo_Municipio' and col != 'Codigo_Municipio_2']] 
```

```python
with pd.ExcelWriter("../dataset/Economico/Renda e Desenvolvimento/PIB/caged_Atualizado.xlsx") as writer: 
    media_estoque_por_ano.to_excel(writer, sheet_name="sheet 1",index=False)
```
